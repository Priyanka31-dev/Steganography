```python
import tkinter as tk
from tkinter import filedialog, messagebox
from PIL import Image

class SteganographyApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Steganography Tool")
        self.root.geometry("700x550")
        self.root.configure(bg="#2e003e")

        # Image path input
        tk.Label(root, text="Image Path", bg="#2e003e", fg="white").pack()
        self.entry_path = tk.Entry(root, width=60)
        self.entry_path.pack()
        tk.Button(root, text="Browse", command=self.browse_image).pack()

        # Message input
        tk.Label(root, text="Message", bg="#2e003e", fg="white").pack()
        self.text_input = tk.Text(root, height=10, width=60)
        self.text_input.pack()

        # Action buttons
        tk.Button(root, text="Encode", command=self.encode_action).pack(pady=5)
        tk.Button(root, text="Decode", command=self.decode_action).pack(pady=5)

    def browse_image(self):
        path = filedialog.askopenfilename(filetypes=[("Image Files", "*.png *.jpg *.jpeg")])
        self.entry_path.delete(0, tk.END)
        self.entry_path.insert(0, path)

    def encode_action(self):
        #encoding logic
        image_path = self.entry_path.get()
        message = self.text_input.get("1.0", tk.END).strip()
        messagebox.showinfo("Stub", f"Encoding message to: {image_path}")
      
        message = self.text_input.get("1.0", tk.END).strip()
        if not image_path or not message:
            messagebox.showwarning("Error", "Please select image and enter a message.")
            return
        output_path = filedialog.asksaveasfilename(defaultextension=".png")
        if output_path:
            success = encode_image(image_path, message, output_path)
            if success:
                messagebox.showinfo("Success", "Message encoded successfully.")
            else:
                messagebox.showerror("Failed", "Message too long or image too small.")

    def decode_action(self):
        #decoding logic
        image_path = self.entry_path.get()
        messagebox.showinfo("Stub", f"Decoding message from: {image_path}")
        if not image_path:
            messagebox.showwarning("Error", "Please select image to decode.")
            return
        result = decode_image(image_path)
        text_input.delete("1.0", tk.END)
        text_input.insert(tk.END, result)

if __name__ == "__main__":
    root = tk.Tk()
    app = SteganographyApp(root)
    root.mainloop()
```
